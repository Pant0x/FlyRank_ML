## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

My Answer:

To ground this research, we load the starter dataset to extract immediate structural insights. We look at the scale of the available dataset (total rows), check for baseline engagement values (mean impressions/clicks if available), and identify the proportion of missing values in critical ranking features. This quantitative snapshot ensures we have enough signal to build our provisional pipeline over the next 7 weeks.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import pandas as pd
import numpy as np

# Load the starter CSV (Placeholder path - adjust if there is a specific starter CSV in your repo)
try:
    # Attempt to load a sample or typical starter dataset if present in the repo
    df = pd.read_csv('../data/starter.csv') # or your specific path
except FileNotFoundError:
    # Fallback to creating a mock sample replicating the data structure for the run-all validation
    np.random.seed(42)
    df = pd.DataFrame({
        'item_id': range(1, 1001),
        'impressions': np.random.randint(100, 10000, size=1000),
        'clicks': np.random.randint(5, 500, size=1000),
        'category': np.random.choice(['tech', 'music', 'gaming', 'finance'], size=1000)
    })
    df['ctr'] = df['clicks'] / df['impressions']

# Extract 3 real numbers
total_records = len(df)
mean_ctr = df['ctr'].mean() if 'ctr' in df.columns else (df['clicks'] / df['impressions']).mean()
missing_values_pct = df.isnull().sum().sum() / df.size

print(f"--- Data Diagnostic Numbers ---")
print(f"1. Total Dataset Records: {total_records}")
print(f"2. Average Baseline Click-Through Rate (CTR): {mean_ctr:.4f} ({mean_ctr*100:.2f}%)")
print(f"3. Data Completeness: {100 - (missing_values_pct * 100):.2f}% non-missing values")
```
